visualizing all the interesting data you can uncover about Snow Geese.

```python
# loading the data
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../data/birds.csv')
birds.head()
```

```python
# Getting the data for snow goose
snow_goose = birds[birds['Name'].str.contains('snow goose', case=False, na=False)]
snow_goose
```

```python
plt.plot(snow_goose['MaxBodyMass'], color='red' , marker='o', label='MaxBodyMass')
plt.plot(snow_goose['MinBodyMass'], marker='o', label='MinBodyMass')
plt.title('Snow Goose Body Mass Range')
plt.show()
```

```python
plt.bar(['Snow Goose'], snow_goose['MaxLength']-snow_goose['MinLength'], bottom=snow_goose['MinLength'])
plt.title('Snow Goose Length Range')
plt.ylim(50,100)
plt.ylabel('Length (cm)')
plt.yticks([50, snow_goose['MinLength'].values[0], snow_goose['MaxLength'].values[0], 100])
plt.show()
```

```python
plt.bar(['Snow Goose'], snow_goose['MaxLength'])
plt.bar(['Snow Goose'], snow_goose['MinLength'])
plt.title('Snow Goose Wingspan Min and Max')
plt.show()
```
