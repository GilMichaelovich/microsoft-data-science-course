# Let's learn about birds


```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../data/birds.csv')
birds.head()
```

```python
counts = birds['Category'].value_counts()
plt.barh(y=counts.index, width=counts.values)
plt.xlabel('counts')
plt.ylabel('Category')
plt.show()
```

```python
plt.barh(y=birds['Family'], width=birds['MaxLength'])
plt.xlabel('MaxLength')
plt.ylabel('Family')
plt.show()
```

```python
grouped = birds.groupby('Category')

MaxBodyMass	= grouped['MaxBodyMass'].max()
MinBodyMass	= grouped['MinBodyMass'].min()
category = MaxBodyMass.index

plt.scatter(MaxBodyMass, category, color='red')
plt.scatter(MinBodyMass, category)
plt.show()
```
