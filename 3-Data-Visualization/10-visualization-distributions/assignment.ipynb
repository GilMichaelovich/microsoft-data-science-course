An athlete dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("athlete_data.csv")
print(df.head())
type(df)
```

1. A histogram of the heights of the athletes in intervals of 10 cm.

```python
df['Height_cm'].plot(kind = 'hist', bins = 5, figsize = (12,12))
plt.show()
```

2. A density plot of the ages of athletes.

```python
sns.kdeplot(df['Age'], bw_adjust=.5)
plt.show()
```

3. A density plot that show the risk of ACL injury per age group.

```python
bins = [17, 20, 30, 40]
labels = ["18-20", "21-30", "31-40"]
df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels)

sns.kdeplot(
   data=df, x="ACL_Risk_Score", hue="Age_Group",
   fill=True, common_norm=False, palette="viridis",
   alpha=.5, linewidth=1,
)
```

4. I made a plot to see if there is any correlation between the amount of training hours per week and the recovery days during the week

```python
sns.kdeplot(data=df, x="Training_Hours_Per_Week", y="Recovery_Days_Per_Week", cut=0, fill=True)
```

5. I want to see which gender has a better team contribution score

```python
sns.kdeplot(data=df, x="Team_Contribution_Score", hue="Gender", multiple="fill")
```
