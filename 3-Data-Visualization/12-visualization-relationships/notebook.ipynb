# Visualizing Honey Production 🍯 🐝

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('../../data/honey.csv')
honey.head()
```

```python
sns.relplot(x="priceperlb", y="state", hue="year", palette=sns.color_palette("tab10"), data=honey, height=15, aspect=.5);
```

```python
sns.relplot(x="priceperlb", y="state", size="year", data=honey, height=15, aspect=.5);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("GDP.csv")
df.columns = df.columns.str.strip()
sample_df = df[df["Country Code"].isin(["USA", "CHN", "JPN", "DEU", "IND", "CAN"])]
sample_df
```

```python
df_long = pd.melt(
    sample_df, 
    id_vars=['Country', 'Country Code'], 
    var_name='Year', 
    value_name='Value'
)
df_long['Year'] = df_long['Year'].astype(int)
df_long = df_long.dropna()
df_long.head()

sns.relplot(
    data=df_long, 
    x="Year", y="Value",
    col="Country", 
    col_wrap=3,
    kind="line"
    )
```
