```python
import pandas as pd

path = '../../data/taxi.csv'

#Load the csv file into a dataframe
df = pd.read_csv(path)

#Print the dataframe
print(df)
```

# Use the cells below to do your own Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

tip_corr_df = pd.DataFrame(
    df.corr(numeric_only=True)["tip_amount"].drop("tip_amount").sort_values()
)

plt.figure(figsize=(4, max(5, len(tip_corr_df) * 0.4)))
sns.heatmap(
    tip_corr_df,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    fmt=".2f",
    cbar_kws={"label": "Correlation Coefficient"}
)

plt.title("tip_amount Correlation Strip")
plt.show()
```

## Q1
From this correlation chart you can infer that there is moderate correlation between the tip amount and the payment type, the total amount of the ride, the trip distance and the tolls amount.
